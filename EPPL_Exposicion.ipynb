#### 🚦 Estructuras Condicionales

Las decisiones en programación se realizan mediante las sentencias:

if

Ejemplo

```python
calificacion = 85

if calificacion >= 90:
    print("Excelente")
elif calificacion >= 70:
    print("Aprobado")
else:
    print("Reprobado")
```

El programa evalúa condiciones de arriba hacia abajo hasta encontrar una verdadera.

#### 🧠 Condicionales Anidados

Permiten tomar decisiones más específicas.

```python
edad = 18
credencial = True

if edad >= 18:
    if credencial:
        print("Puede votar")
    else:
        print("Necesita identificación")
else:
    print("Menor de edad")
```

#### 🚀 TensorFlow en Google Colab

TensorFlow es una biblioteca desarrollada para crear modelos de Inteligencia Artificial y Aprendizaje Automático.

Google Colab permite ejecutarla directamente desde el navegador sin instalar software.

```python
import tensorflow as tf
import numpy as np

entrada = np.array([1, 2, 3, 4, 5], dtype=float)
salida = np.array([2, 4, 6, 8, 10], dtype=float)

modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

modelo.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)

modelo.fit(
    entrada,
    salida,
    epochs=500,
    verbose=0
)

print(modelo.predict(np.array([7.0])))
```

#### 📈 ¿Qué Hace Este Programa?
Se proporcionan ejemplos de entrada y salida.

La red neuronal analiza los patrones.
Aprende que la salida es el doble de la entrada.

Al ingresar un nuevo valor, realiza una predicción.

Ejemplo

```python
Entrada: 7
Predicción: 14
```
